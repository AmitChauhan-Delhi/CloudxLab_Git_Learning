```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
norm_data = pd.DataFrame(np.random.normal(size=10000))
norm_data.plot.density()
# norm_data.plot(kind="density", figsize=(5,5))
plt.axvline(norm_data[0].mean(), color='red', linestyle='dashed', label='Mean')
plt.axvline(norm_data[0].median(), color='green', linestyle='dashed', label='Median')
plt.xlabel('Data')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
skewed_data = pd.DataFrame(np.random.exponential(size=100000))
skewed_data.plot(kind="density",figsize=(5,5),xlim=(-1,5))
plt.axvline(norm_data[0].mean(), color='red', linestyle='dashed', label='Mean')
plt.axvline(norm_data[0].median(), color='green', linestyle='dashed', label='Median')
plt.legend()
plt.show()
```
